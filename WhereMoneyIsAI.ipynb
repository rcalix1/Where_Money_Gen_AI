```python

import replicate

input = {
    "top_p": 1,
    "prompt": "Tell me how to tailor a men's suit so I look fashionable.",
    "temperature": 0.75,
    "max_new_tokens": 800
}

for event in replicate.stream(
    "meta/llama-2-7b-chat",
    input=input
):
    print(event, end="")
#=> " Of course, I'd be happy to help! Tailoring a men's suit...
```

```python

import torch
## torch.cuda.empty_cache()

from fastai.text.all import *
```

```python

dls = TextDataLoaders.from_folder(
           untar_data(URLs.IMDB), bs=8, valid='test'
)

learn = text_classifier_learner(dls, AWD_LSTM, drop_mult=0.5, metrics=accuracy)

learn.fine_tune(1, 1e-2)

learn.predict("I really do not like the movie the matrix!")
```
